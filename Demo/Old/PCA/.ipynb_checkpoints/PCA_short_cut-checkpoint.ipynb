```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA as sklearnPCA
import matplotlib.pyplot as plt


table = pd.read_csv("data/heart_table.csv")
table
```

```python

X_std = StandardScaler().fit_transform(table)
```

```python

sklearn_pca = sklearnPCA(.80)
X_pca = sklearn_pca.fit_transform(X_std)
```

```python

pc1 = X_pca[:,0]
pc2 = X_pca[:,1]
diag = X_std[:,-1]

plt.scatter(pc1, pc2,c=diag, cmap=plt.cm.Paired)
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.show()
```
