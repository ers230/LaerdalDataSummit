```python
# https://jakevdp.github.io/PythonDataScienceHandbook/05.07-support-vector-machines.html

import numpy as np
from sklearn import datasets, svm
import matplotlib.pyplot as plt
import pandas as pd


%matplotlib inline
```

```python
iris = datasets.load_iris()
classes = {}
for i in range(0,len(iris.target_names)):
    classes[i] = iris.target_names[i]

def load_iris_dataset():
    X = iris.data[:, 2:]
    y = iris.target
    
    y_names = [classes[i] for i in y]
    
    df = pd.DataFrame({'petal_length': X[:,0], 'petal_width': X[:,1], 'target_values': y, 'target_names': y_names})
    df = df[['petal_length', 'petal_width', 'target_values', 'target_names']]
    return df
```

```python
classifiers = ['SVC with linear kernel',
              'LinearSVC (linear kernel)',
              'SVC with RBF kernel',
              'SVC with polynomial (degree 3) kernel']

#@fusion.register(args={'classifier':{"values": classifiers}})
def SVM_classifiers(X=None, y=None, classifier='SVC with linear kernel'):
    """
    Using SVM Classifiers function
    ------------------------------

    The clustering function receives a 2-colummns table (x, y) `data`
    and applies the selected `classifier` to the dataset

    The available algorithms are:

    * SVC with linear kernel
    * LinearSVC (linear kernel)
    * SVC with RBF kernel
    * SVC with polynomial (degree 3) kernel

    [Click here for more information about SVM classifiers and scikit-learn](http://scikit-learn.org/stable/auto_examples/svm/plot_iris.html)
    """
    
    # import some data to play with
    iris = datasets.load_iris()
    if X is None: 
        X = iris.data[:, 2:]  # we only take the first two features. We could
                          # avoid this ugly slicing by using a two-dim dataset
    else:
        X = np.array(X)
        
    if y is None:
        y = iris.target
    else:
        y = np.asarray(y).flatten()
        
    h = .02  # step size in the mesh

    # we create an instance of SVM and fit out data. We do not scale our
    # data since we want to plot the support vectors
    C = 1.0  # SVM regularization parameter
    
    if classifier=='SVC with linear kernel':
        clf = svc = svm.SVC(kernel='linear', C=C).fit(X, y)
    elif classifier=='SVC with RBF kernel':
        clf = svm.SVC(kernel='rbf', gamma=0.7, C=C).fit(X, y)
    elif classifier=='SVC with polynomial (degree 3) kernel':
        clf = svm.SVC(kernel='poly', degree=3, C=C).fit(X, y)
    elif classifier=='LinearSVC (linear kernel)':
        clf = svm.LinearSVC(C=C).fit(X, y)
    else:
        clf = svc = svm.SVC(kernel='linear', C=C).fit(X, y)
    
    # create a mesh to plot in
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                         np.arange(y_min, y_max, h))

    Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])

    # Put the result into a color plot
    Z = Z.reshape(xx.shape)

    plt.contourf(xx, yy, Z, cmap=plt.cm.Paired, alpha=0.8)

    # Plot also the training points
    plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.Paired)
    plt.xlabel('Petal length')
    plt.ylabel('Petal width')
    plt.xlim(xx.min(), xx.max())
    plt.ylim(yy.min(), yy.max())
    plt.xticks(())
    plt.yticks(())
    
    plt.show()
    
    predicted = clf.predict(X).tolist()
    predicted_names = [classes[i] for i in predicted]
    
    df = pd.DataFrame({'predicted_values': predicted, 'predicted_names': predicted_names})
    df = df[['predicted_values', 'predicted_names']]
    
    return df
```
